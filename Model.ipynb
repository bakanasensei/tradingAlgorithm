```python
trainData = np.array(df)
trainData[0]
```

```python
labelData = np.array(dfLabels)
testData = trainData[15000:]
testLabels = labelData[15000:]
train = trainData[:15000]
labels = labelData[:15000]
```

```python
model = Sequential([
    Dense(units=80, input_shape=(100,), activation='relu'),
    Dense(units=160, activation='relu'),
    Dense(units=60, activation='relu'),
    Dense(units=20, activation='relu'),
    Dense(units=2, activation='softmax')
])
model.summary()
```

```python
class MulticlassTruePositives(tf.keras.metrics.Metric):
    def __init__(self, name='multiclass_true_positives', **kwargs):
        super(MulticlassTruePositives, self).__init__(name=name, **kwargs)
        self.true_positives = self.add_weight(name='tp', initializer='zeros')

    def update_state(self, y_true, y_pred, sample_weight=None):
        y_pred = tf.reshape(tf.argmax(y_pred, axis=1), shape=(-1, 1))
        values = tf.cast(y_true, 'int32') == tf.cast(y_pred, 'int32')
        values = tf.cast(values, 'float32')
        if sample_weight is not None:
            sample_weight = tf.cast(sample_weight, 'float32')
            values = tf.multiply(values, sample_weight)
        self.true_positives.assign_add(tf.reduce_sum(values))

    def result(self):
        return self.true_positives

    def reset_states(self):
        # The state of the metric will be reset at the start of each epoch.
        self.true_positives.assign(0.)
        

model.compile(optimizer=Adam(learning_rate=0.00001), loss='sparse_categorical_crossentropy', metrics=['accuracy', MulticlassTruePositives()])
model.fit(x=train, y=labels, batch_size=10, epochs=100, verbose=2)
```

```python
predictions = model.predict(
      x=testData
    , batch_size=10
    , verbose=0
)
for i in predictions:
    print(i)
```

```python
rounded_predictions = np.argmax(predictions, axis=-1)

ok = 0
for i in range(rounded_predictions.shape[0]):
    print(str(rounded_predictions[i]) + ' ' + str(testLabels[i][0]))
    if rounded_predictions[i] == testLabels[i][0]:
        ok += 1

tp = ok / testLabels.shape[0]
print('\n')
print(tp)
```

```python
strongPredictions = list()
for elem in predictions:
    if (elem[0] >= 0.8) or (elem[1] >= 0.8):
        strongPredictions.append(elem)
        
print('Strong prediction percentage: ' + str(len(strongPredictions) / testLabels.shape[0]))
print('Strong predictions: ' + str(len(strongPredictions)))
```
